# Bibliotecas

```python
from utils import carregar_dados, preprocessar
from genetic import algoritmo_genetico

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    recall_score,
    precision_score,
    roc_auc_score
)

import numpy as np
import pandas as pd
```

Carregar Dados

```python
df = carregar_dados()
X_tr, X_te, y_tr, y_te = preprocessar(df)
```

Parametros


```python
GA_CONFIG = {
    "pop_size": 20,
    "generations": 8,
    "mutation_rate": 0.40,
    "tournament_size": 4,
    "penalizacao": 0.04
}
```

baseline

```python
resultados = []
```

```python
for i in range(10):

    best_ind, score = algoritmo_genetico(
        X_tr, y_tr, X_te, y_te, GA_CONFIG
    )

    # seleção de features
    X_tr_sel = X_tr[:, best_ind == 1]
    X_te_sel = X_te[:, best_ind == 1]

    # modelo final
    model = GaussianNB()
    model.fit(X_tr_sel, y_tr)

    pred = model.predict(X_te_sel)

    # métricas
    acc = accuracy_score(y_te, pred)

    f1 = f1_score(y_te, pred, average=None)
    recall = recall_score(y_te, pred, average=None)
    precision = precision_score(y_te, pred, average=None)

    auc = roc_auc_score(y_te, pred)

    resultados.append([
        i + 1,
        acc,
        recall[1],
        precision[1],
        f1[1],
        recall[0],
        precision[0],
        f1[0],
        auc,
        np.sum(best_ind),
        len(X_tr_sel)
    ])
```

Resultados

```python
colunas = [
    "Execução",
    "Acc",
    "R+",
    "P+",
    "F1+",
    "R-",
    "P-",
    "F1-",
    "AUC-ROC",
    "Variáveis",
    "Instâncias de Treinamento"
]

df_resultados = pd.DataFrame(resultados, columns=colunas)

df_resultados
```
