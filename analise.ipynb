# Bibliotecas

```python
from utils import carregar_dados, preprocessar
from genetic import algoritmo_genetico
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
```

Carregar Dados

```python
df = carregar_dados()
X_tr, X_te, y_tr, y_te = preprocessar(df)
```

Parametros


```python
GA_CONFIG = {
    "pop_size": 50,
    "generations": 80,
    "mutation_rate": 0.1,
    "tournament_size": 3,
    "penalizacao": 0.01
}
```

baseline

```python
model = GaussianNB()
model.fit(X_tr, y_tr)

pred = model.predict(X_te)

print("BASELINE")
print("Accuracy:", accuracy_score(y_te, pred))
print(classification_report(y_te, pred))
```

GA

```python
best_ind, score_ga = algoritmo_genetico(
    X_tr,
    y_tr,
    X_te,
    y_te,
    GA_CONFIG
)

print("GA SCORE:", score_ga)
print("Features:", np.sum(best_ind))
```

modelo final

```python
X_tr_sel = X_tr[:, best_ind == 1]
X_te_sel = X_te[:, best_ind == 1]

model = GaussianNB()
model.fit(X_tr_sel, y_tr)

pred = model.predict(X_te_sel)

print("GA FINAL")
print("Accuracy:", accuracy_score(y_te, pred))
print(classification_report(y_te, pred))
```
